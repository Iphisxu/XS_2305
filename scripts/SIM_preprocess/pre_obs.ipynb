```python
import xarray as xr
import numpy as np
import pandas as pd

import sys
sys.path.append('../../src/')
from namelist import *

# silence the warning note
import warnings
warnings.filterwarnings("ignore")
```

```python
# 读取原始数据
xls=pd.read_excel('D:/data/Project_Xiaoshan/TEST_SIM/obsdata/originaldata.xlsx',header=0,na_values=['—',-99])
xls
```

```python

# 将日期与时间列合并
xls['datetime'] = pd.to_datetime(xls['Year'].astype(str)+'-'+xls['Month'].astype(str)+'-'+xls['Day'].astype(str)+' '+xls['Hour'].apply(lambda x: '{0:0>2}'.format(x)))

# 删除“日期”列和“时间”列
xls.drop(['Year','Month','Day','Hour'], axis=1, inplace=True)

# 设置为索引
xls.set_index('datetime',inplace=True)
xls
```

```python
# 将所有因包含括号等字符被读取为字符串的数据设置为空值
xls = xls.replace(to_replace=[r'\(.*?\)'], value=np.nan, regex=True)
```

```python
sites = pd.read_excel('D:/data/Project_Xiaoshan/TEST_SIM/obsdata/siteslocation.xlsx')
sites
```

```python
# 根据站点分组导出
grouped = xls.groupby('Station')
i=0
for group_name, group_df in grouped:
    filename = str(sites['城市'][i]+'-'+sites['站点'][i])
    group_df.to_excel(f'D:/Download/{filename}.xlsx', index=True)
    i+=1
```

```python
# 将全部站点取平均并导出
xls.groupby(xls.index).mean().to_excel('D:/Download/avg_2306.xlsx',index=True)
```
